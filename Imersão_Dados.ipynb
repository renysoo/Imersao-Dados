##Desafio 01: Ordenar o gráfico countplot


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))

dadosRelevantes = dados.query('composto in @cod_compostos')

ax = sns.countplot(x = 'composto', data = dadosRelevantes, order=dadosRelevantes['composto'].value_counts().index)
ax.set_title('Top 5 compostos')
plt.show()
```

##Desafio 02: Melhorar a visualização alterando tamanho da fonte...

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))

dadosRelevantes = dados.query('composto in @cod_compostos')

ax = sns.countplot(x = 'composto', data = dadosRelevantes, order=dadosRelevantes['composto'].value_counts().index, palette='Set3')
ax.set_title('Top 5 compostos',fontweight='bold', fontsize=20)
ax.set_xlabel('Composto',fontsize=15, labelpad=10)
ax.set_ylabel('Contagem', fontsize=15, labelpad=10)
for patch in ax.patches:
  ax.annotate(patch.get_height(), (patch.get_x()+0.25, patch.get_height()))
plt.show()
```

##Desafio 03: Plotar os histogramas com seaborn


```python
plt.figure(figsize=(8,6))
dadosHist = dados
ax = sns.histplot(x='g-0',data=dadosHist,kde=True)
ax.set_title('Resultados g-0',fontsize=20,fontweight='bold')
ax.set_xlabel('valores',fontsize=10, labelpad=10)
ax.set_ylabel('quantidade',fontsize=10, labelpad=10)
plt.show()
```

##Desafio 04: Estudar sobre as estatísticas retornadas no .describe()


```python
dados.loc[:,'g-0':'g-771'].describe().T
```

```python
plt.figure(figsize=(6,3))
dadosHist = dados.loc[:,'g-0':'g-771'].describe().T
ax = sns.histplot(x='mean',data=dadosHist)
ax.set_title('Média',fontsize=20,fontweight='bold')
ax.set_xlabel('valores',fontsize=10, labelpad=10)
ax.set_ylabel('quantidade',fontsize=10, labelpad=10)

plt.figure(figsize=(6,3))
dadosHist = dados.loc[:,'g-0':'g-771'].describe().T
ax = sns.histplot(x='std',data=dadosHist)
ax.set_title('Desvio padrão',fontsize=20,fontweight='bold')
ax.set_xlabel('valores',fontsize=10, labelpad=10)
ax.set_ylabel('quantidade',fontsize=10, labelpad=10)

plt.figure(figsize=(6,3))
dadosHist = dados.loc[:,'g-0':'g-771'].describe().T
ax = sns.histplot(x='min',data=dadosHist)
ax.set_title('Valor mínimo',fontsize=20,fontweight='bold')
ax.set_xlabel('valores',fontsize=10, labelpad=10)
ax.set_ylabel('quantidade',fontsize=10, labelpad=10)

plt.figure(figsize=(6,3))
dadosHist = dados.loc[:,'g-0':'g-771'].describe().T
ax = sns.histplot(x='max',data=dadosHist)
ax.set_title('Valor máximo',fontsize=20,fontweight='bold')
ax.set_xlabel('valores',fontsize=10, labelpad=10)
ax.set_ylabel('quantidade',fontsize=10, labelpad=10)

plt.figure(figsize=(6,3))
dadosHist = dados.loc[:,'g-0':'g-771'].describe().T
ax = sns.histplot(x='50%',data=dadosHist)
ax.set_title('Mediana',fontsize=20,fontweight='bold')
ax.set_xlabel('valores',fontsize=10, labelpad=10)
ax.set_ylabel('quantidade',fontsize=10, labelpad=10)

plt.show()
```
